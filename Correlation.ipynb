```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


# 导入数据
data = pd.read_csv('cleaned_file.csv')
df = pd.DataFrame(data, columns=['Voltage', 'multiple', 'k', 'rpm', 'P', 'I', 'Stop Duration'])

# 计算相关性矩阵
corr_matrix = df.corr()

# 使用 seaborn 绘制热力图
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, vmin=-1, vmax=1)

# 添加标题
plt.title('Correlation Heatmap')

# 显示图形
plt.show()
```
